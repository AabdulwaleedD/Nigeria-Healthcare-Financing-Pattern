Standardize Column Names
# Renaming Columns names to be specific to the file content

```python
total_spend = total_spend.rename(columns={'NumericValue': 'Total_USD_pc', 'TimeDim': 'Year', 'SPEND':'SpatialDimension','ParentLocationCode':'Continent_Code',
                                         'ParentLocation':'Continent', 'SpatialDimensionValueCode':'Country_Code'})
govt_spend = govt_spend.rename(columns={'NumericValue': 'Govt_USD_pc', 'TimeDim': 'Year','ParentLocationCode':'Continent_Code',
                                         'ParentLocation':'Continent', 'SpatialDimensionValueCode':'Country_Code'})
oop_spend = oop_spend.rename(columns={'NumericValue': 'OOP_USD_pc', 'TimeDim': 'Year','ParentLocationCode':'Continent_Code',
                                         'ParentLocation':'Continent', 'SpatialDimensionValueCode':'Country_Code'})
```

Merge into one Master DataFrame

```python
df = pd.merge(total_spend[['Continent','Country_Code','Year', 'Total_USD_pc']], 
              govt_spend[['Year', 'Govt_USD_pc']], on='Year', how='outer')
df = pd.merge(df, oop_spend[['Year', 'OOP_USD_pc']], on='Year', how='outer')
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['Year'], df['Total_USD_pc'], label='Total Spending', marker='o')
plt.plot(df['Year'], df['Govt_USD_pc'], label='Govt Spending', marker='s')
plt.plot(df['Year'], df['OOP_USD_pc'], label='Out-of-Pocket', marker='^')

plt.title('Nigeria Healthcare Spending Trends (USD per capita)')
plt.xlabel('Year')
plt.ylabel('USD ($)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Set the style for a professional look
sns.set_theme(style="whitegrid")
plt.figure(figsize=(12, 7))

# Plotting the three core metrics from your merged dataframe (df)
# Blue for Total, Red for Government, Orange/Green for Out-of-Pocket
sns.lineplot(data=df, x='Year', y='Total_USD_pc', label='Total Health Spending', color='#2c3e50', linewidth=3, marker='o')
sns.lineplot(data=df, x='Year', y='Govt_USD_pc', label='Govt Health Spending', color='#e74c3c', linewidth=3, marker='s')
sns.lineplot(data=df, x='Year', y='OOP_USD_pc', label='Out-of-Pocket Spending (Household)', color='#27ae60', linewidth=2, linestyle='--', marker='^')

# Highlighting the "Funding Gap"
plt.fill_between(df['Year'], df['Govt_USD_pc'], df['Total_USD_pc'], color='gray', alpha=0.1, label='The Funding Gap')

# Chart Details
plt.title('Nigeria Health Financing: The Growing Gap (2000 - 2023)', fontsize=16, fontweight='bold', pad=20)
plt.ylabel('USD ($) per capita', fontsize=12)
plt.xlabel('Year', fontsize=12)
plt.legend(loc='upper left', frameon=True)

# Add an annotation for the "Abuja Declaration" context
plt.annotate('Abuja Declaration (2001)\nPledged 15% Budget', 
             xy=(2001, 17), xytext=(2003, 40),
             arrowprops=dict(facecolor='black', shrink=0.05, width=1, headwidth=5))

plt.tight_layout()
plt.savefig('Nigeria_Health_Finance_Story.png', dpi=300)
plt.show()
```
